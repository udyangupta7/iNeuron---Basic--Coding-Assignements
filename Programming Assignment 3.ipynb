Q1: Write a Python Program to Check if a Number is Positive, Negative or Zero?

A1:

```python
a = int(input("Enter a number, negative or positive:- "))
if a > 0:
    print(a,"is a positive number.")
if a == 0:
    print("The number entered is zero.")
if a < 0:
    print(a,"is a negative number.")
```

Q2: Write a Python Program to Check if a Number is Odd or Even?
    
A2:

```python
b = int(input("Enter a number:- "))
if b % 2 == 0:
    print(b,"is an even number.")
else:
    print(b,"is an odd number.")
```

Q3: Write a Python Program to Check Leap Year?
    
A3:

```python
c = int(input("Enter a number:- "))
if c % 4 == 0:
    print(c,"is a leap year.")
else:
    print(c,"is a non-leap year.")
```

Q4: Write a Python Program to Check Prime Number?
    
A4:

```python
d = int(input("Enter a number:- "))
f = False
if d > 1:
    for i in range(2, d):
        if (d % i) == 0:
            f = True
            break
if f:
    print(d,"is not a prime number")
else:
    print(d,"is a prime number")
```

Q5: Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

A5:

```python
for num in range(1,10001):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num,"\t",end=" ")
```
